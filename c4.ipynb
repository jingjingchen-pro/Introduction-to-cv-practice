```python
from math import sqrt
import matplotlib.pyplot as plt
```

```python
X1 = 5
X2 = -5
f = 1
```

```python
L = []
for Z in range(10, 1000):
    x1 = -f * X1 / Z
    x2 = -f * X2 / Z
    L.append( sqrt((x1 - x2) ** 2))
```

```python
Z_values = list(range(10, 1000))
plt.plot(Z_values, L)
plt.xlabel('distance(Z)')
plt.ylabel('length')
```
